```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import operator
print(f"open cv ver: {cv2.__version__}")
```

```python
plt.imshow(cv2.imread('fav_photo.jpg'))
```

```python
lab_image = cv2.cvtColor(cv2.imread('fav_photo.jpg'), cv2.COLOR_RGB2LAB)
plt.imshow(lab_image)
```

```python
def playWithChannels(ChannelA, ChannelB, op1, op2, ValueA, ValueB):
    ChannelA = op1(ChannelA, ValueA)
    ChannelB = op2(ChannelB, ValueB)
    return ChannelA, ChannelB
```

```python
L, A, B = cv2.split(lab_image)
```

```python
list_of_operators = [operator.add, operator.sub]
list_of_experiments = []
for i in range(10):
    random_op_one = np.random.choice(list_of_operators)
    random_op_two = np.random.choice(list_of_operators)
    temp_A, temp_B = playWithChannels(ChannelA=A, ChannelB=B, op1=np.random.choice(list_of_operators), op2=np.random.choice(list_of_operators), ValueA=np.random.random_integers(0, 20), ValueB=np.random.random_integers(0, 20)  )
    list_of_experiments.append(cv2.merge([L, temp_A, temp_B]))
```

```python
rows = 5
columns = 2
fig = plt.figure(figsize=(20, 20))
j = 1
for i in list_of_experiments:
    fig.add_subplot(rows, columns, j)
    j+=1
    plt.imshow(cv2.cvtColor(i, cv2.COLOR_LAB2RGB))
plt.show()
```
