```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
print(f"open cv2 ver: {cv2.__version__}")
```

```python
clf = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
image = cv2.cvtColor(cv2.imread('billy.jpg', cv2.COLOR_BGR2RGB), cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image)
```

```python
face = clf.detectMultiScale(gray_image, 1.1, 6)
for (x, y, w, h) in face:
    cv2.rectangle(image, (x,y), (x+w, y+h), (255, 0, 0), 2)
    # roi_gray = gray_image[y:y+h, x:x+w]
    # roi_color = image[y:y+h, x:x+w]

plt.imshow(image)
```
